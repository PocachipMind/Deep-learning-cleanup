## Line Plot

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_kospi = pd.read_csv('../Data/KOSPI.csv', index_col = 'Date')
df_kospi
````

````python
df_kospi.plot()
````

````python
plt.figure() #도화지를 생성

df_kospi.plot(figsize = (15,4))

plt.axhline(2000, linestyle = '--', color = 'red') # 라인 그리기

plt.show()
````

````python
fig, ax = plt.subplots()

df_kospi.plot(kind = 'line', ax = ax, figsize = (15,10))

ax.axhline(2000, linestyle = '--', color = 'green')

ax.set(xlabel = "Date", ylabel = 'KOSPI', title = 'Time Series of Kospi Index')

plt.show()
````

## 두개 이상의 그래프

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_samsung = pd.read_excel('../Data/samsung_naver_excel.xlsx', sheet_name=0, index_col = 'Date')
df_samsung
````

````python
fig, axs = plt.subplots(2,1, figsize = (15,10)) # 도화지 준비

axs[0].plot(df_samsung['Close']['2018-05-04':], color = 'blue')

axs[0].grid(True)

axs[0].set(xlabel = "Date", ylabel = 'Price', title = 'Time Series of Samsung Price')

axs[0].plot(df_samsung['Open']['2018-05-04':], color = 'red')

axs[1].scatter(df_samsung.index, df_samsung['Volume'], color = 'red')

axs[1].grid(True)

axs[1].set(xlabel = "Date", ylabel = 'Volume')

plt.show()
````

````python
fig, axs = plt.subplots(3,2, figsize = (15,10)) # 도화지 준비

axs[0,0].plot(df_samsung['Close']['2018-05-04':], color = 'blue')

axs[0,0].grid(True)

axs[0,0].set(xlabel = "Date", ylabel = 'Price', title = 'Time Series of Samsung Price')

axs[0,0].plot(df_samsung['Open']['2018-05-04':], color = 'red')

axs[0,1].scatter(df_samsung.index, df_samsung['Volume'], color = 'red')

axs[0,1].grid(True)

axs[0,1].set(xlabel = "Date", ylabel = 'Volume')

plt.show()
````

## Option

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 그림 설정
plt.rcParams["figure.figsize"] = (22,15) # 크기
plt.rc('font', family = 'Malgun Gothic') # 한글 쓸 수 있는 폰트 맑음고딕
plt.rcParams["font.size"] = 15 # 폰트 크기
plt.rc('axes', unicode_minus = False) # 마이너스 부분 폰트 

# 데이터 불러오기
df = pd.read_excel('../Data/samsung_kospi_excel.xlsx', sheet_name=0, index_col = 'Date')
df = df.loc['2018-01-02':]
df
````

````python
# 기본 베이스 코드

fig, axes = plt.subplots(3,1)

axes[0].plot(df['Close'], color = 'red')
axes[0].set_title("Samsung Stock")

axes[1].plot(df['Close'], color = 'blue')
axes[1].set_title("수익률")

axes[2].plot(df['Close'], color = 'green')
axes[2].set_title("거래량")

plt.show()
````

````python
fig, axes = plt.subplots(3,1, sharex = False) # sharex 가 True가 되면 인덱스 공유

axes[0].plot(df['Close'], color = 'red', linewidth = 2)
axes[0].plot(df['Close'].rolling(window = 30).mean(), color = 'blue', linewidth = 3)
axes[0].axhline(df['Close'].mean(), linestyle = '--', linewidth = 5) # 라인 그리기
axes[0].scatter(df.index[df['Close'].argmax()], df['Close'].max(), marker = 'o',s = 500)
axes[0].legend(['Close', 'Rolling Mean 30days'])
axes[0].set_title("Samsung Stock")

axes[1].plot(df['Change']*100, color ='blue', linewidth = 0.8)
axes[1].axhline(0, linestyle = '--', linewidth = 4, color = 'red')
axes[1].tick_params(axis = 'x', labelrotation = 45) # 45도 글자 기울이기
axes[1].set_title("수익률")

axes[2].fill_between(df.index, df['Volume'], color = 'green', alpha = 1) # x축, y축, 색상, 투명도
axes[2].set_title("거래량")
axes[2].axvline(df.index[df['Volume'].argmax()], color = 'red', linewidth = 5, linestyle = '--') # 수직선 

plt.tight_layout() # 글자 안곂치게 
plt.show()
````
